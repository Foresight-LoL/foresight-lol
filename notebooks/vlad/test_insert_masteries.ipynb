```python
import polars as pl

df =  pl.read_database_uri(
    query= "SELECT puuid FROM players",
    uri = POSTGRES_LOCAL_URL,
)
df
```

```python
puuid_list = df["puuid"].to_list()[:10]
puuid_list
```

```python
json_collection = []

for puuid in puuid_list:
    url = f"https://eun1.api.riotgames.com/lol/champion-mastery/v4/champion-masteries/by-puuid/{puuid}"
    response = requests.get(url, headers=headers)
    json_collection.extend(response.json())
#df = pl.DataFrame(json_collection)
#df
json_collection
```

```python
from datetime import datetime,timezone
final_mastery_rows = []
def transform_row(row):
    return {
        "puuid": str(row["puuid"]),
        "champion_id": int(row["championId"]),
        "milestone_grades": row.get("milestoneGrades",[]),
        "tokens_earned": row["tokensEarned"],
        "champion_season_milestone": row["championSeasonMilestone"],
        "mark_required_for_next_level": row["markRequiredForNextLevel"],
        "champion_points_since_last_level": row["championPointsSinceLastLevel"],
        "champion_points": row["championPoints"],
        "champion_level": row["championLevel"],
        "last_play_time": datetime.fromtimestamp(row["lastPlayTime"] // 1000),
        "chest_granted": row.get("chestGranted"),#TODO: this is always null, potential fix SCHEMA
        "champion_points_until_next_level": row["championPointsUntilNextLevel"],
    }
for row in json_collection:
    final_mastery_rows.append(transform_row(row))
mastery_df = pl.DataFrame(final_mastery_rows)
mastery_df
```

```python
connection_wrapper.upsert_dataframe_dto_safe(mastery_df, "masteries", ["puuid", "champion_id"])
```
